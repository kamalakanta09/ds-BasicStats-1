```python
# Data Visualization
# Plot histograms for each numerical column
numerical_columns.hist(figsize=(10,12))
plt.suptitle('Histograms for Numerical Columns')
plt.show()
```

```python
# Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.
def plot_histograms(numerical_columns):
  sns.set(style='whitegrid')
  # Define the number of columns for subplot
  num_cols = len(numerical_columns.columns)
  # Define the number of rows based on the number of columns
  num_rows = (num_cols // 2) + (num_cols % 2)

  # Create subplots
  fig, axes = plt.subplots(num_rows, 2, figsize=(15, 2*num_rows))
  fig.subplots_adjust(hspace=0.9)
  # Flatten the axes if the number of rows is greater than 1
  if num_rows > 1:
    axes = axes.flatten()
  else:
    [axes]
  # Plot histograms for each numerical column
  for i, col in enumerate(numerical_columns.columns):
    ax = axes[i]
    sns.histplot(df[col], ax=ax, kde=True)
    ax.set_title(f'Histogram of {col}')
    # Add annotations for skewness and outliers
    skewness = numerical_columns[col].skew()
    ax.annotate(f'Skewness: {skewness:.2f}', xy=(0.7, 0.9), xycoords='axes fraction', fontsize=10)

  # Remove unused subplots if the number of columns is odd
  if num_cols % 2 != 0:
    fig.plt.delaxes(axes[-1])

  plt.show()

# Call the function to plot histograms
plot_histograms(numerical_columns)
```

```python
# Boxplots
numerical_columns.plot(kind='box', subplots=True, layout=(3,4), figsize=(17, 14))
plt.suptitle('Boxplots for Numerical Columns')
plt.show()
```

```python
# Boxplots , not given manually layout
for i in dict(numerical_columns.dtypes):
  if dict(numerical_columns.dtypes)[i] == 'int64' or dict(numerical_columns.dtypes)[i] == 'float64':
    plt.boxplot(numerical_columns[i], vert=False)
    plt.title("Boxplots for Numerical Columns")
    plt.xlabel(i)
    plt.show()
```

```python
# Plot boxplots for each numerical column
for col in numerical_columns:
  plt.figure(figsize=(8, 6))
  plt.title(f'Boxplot of {col}')
  plt.boxplot(df[col])
  plt.xlabel(col)
  plt.ylabel('Values')
  plt.grid(True)
  plt.show()
  # Analyze outliers and unusual distributions
  # Calculate quartiles and IQR
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1 #Interquartile Range (IQR)
  # Calculate upper and lower bounds for outliers
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 - 1.5 * IQR
  # Get the outliers
  outliers = df[(df[col] < lower_bound) | (df[col] < upper_bound)]

  if len(outliers) > 0:
    print(f'Column: {col}')
    print(f'Number of outliers: {len(outliers)}')
    print(f'Outlier values: {outliers[col].tolist()}')
    print('\n')
# Discuss any findings, such as extreme values or unusual distributions
print("Analysis of Boxplots:")
print("1. Outliers: Identify and discuss any outliers found in the data.")
print("2. Skewness and Symmetry: Assess the skewness and symmetry of each boxplot.")
print("3. Unusual Distributions: Describe any patterns or shapes in the boxplots that deviate from expectations.")
print("4. Context: Interpret the findings in the context of your dataset and research question.")
print("5. Implications: Discuss the practical implications of the findings and how they may affect your analysis.")
```

```python
# Categorical Columns Data:
categorical_columns = df.select_dtypes(include='object')
categorical_columns
```

```python
# Bar Chart Analysis for Categorical Column:
for column in categorical_columns:
  plt.figure(figsize=(8,6))
  df[column].value_counts().plot(kind='bar')
  plt.title(f'Bar Chart for {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

```python
# Analyze the distribution of categories and provide insights.
for col in categorical_columns:
   # Count the frequency of each category
   categorical_counts = df[col].value_counts()
   # Create a bar chart
   plt.figure(figsize=(8,6))
   categorical_counts.plot(kind='bar', color='skyblue')
   plt.title(f'Frequency Distribution of Categories in {col}')
   plt.xlabel('Categories')
   plt.ylabel('Frequency')
   plt.xticks(rotation=45) # Rotate x-axis labels for better readability
   plt.grid(axis='y', linestyle='--', alpha=0.7)
   plt.tight_layout()
   plt.show()

   # Provide insights
   print(f"Insights for {col}:")
   print(f"The bar chart above shows the distribution of categories in the '{col}' column.")
   print("Analyzing this distribution helps in understanding the relative frequencies of different categories.")

```

```python
# Standardization of Numerical Variables
# Display summary statistics before normalization
print("Summary Statistics Before Normalization:")
numerical_columns.describe()
```

```python
# Standardization of Numerical Variables
# Perform z-score normalization
standardized_data = (numerical_columns - numerical_columns.mean()) / numerical_columns.std()
# Display summary statistics after normalization
print("\nSummary Statistics After Normalization:")
standardized_data.describe()
```

```python
# Plot histograms for both original and normalized data
fig, axes = plt.subplots(nrows=2, ncols=len(numerical_columns.columns), figsize=(15, 8))

for i, col in enumerate(numerical_columns.columns):
  numerical_columns[col].plot(kind='hist', ax=axes[0, i], title=col + '(Original)')
  standardized_data[col].plot(kind='hist', ax=axes[1, i], title=col + '(Normalized)')

plt.tight_layout()
plt.show()
```

```python
# Create pivot maps for both original and normalized data
pivoteBefore = numerical_columns.melt(var_name='Column', value_name='Value')
pivoteAfter = standardized_data.melt(var_name='Column', value_name='Value')
# Plot pivot maps
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

pivoteBefore.plot.density(ax=axes[0], label='Before Normalization', color='blue')
pivoteAfter.plot.density(ax=axes[1], label='After Normalization', color='orange')

axes[0].set_title('Density Plot Before Normalization')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')

axes[1].set_title('Density Plot After Normalization')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Concatenate original and normalized data
concatenated_data = pd.concat([numerical_columns, standardized_data], axis=1)

# Plot density plots for both original and normalized data on the same plot
fig, axes = plt.subplots(nrows=1, ncols=len(numerical_columns.columns), figsize=(15, 5))

for i, col in enumerate(numerical_columns.columns):
    concatenated_data[col].plot(kind='density', ax=axes[i], title=col)

plt.tight_layout()
plt.show()
```
